```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = [1990, 1992, 1994, 1996, 1998, 2000, 2002, 2004, 2006, 2008, 2010, 2012, 2014]
virat = [123, 34, 23, 45, 67, 84, 36, 35, 56, 43, 23, 52, 60]
sehwag = [12, 32, 54, 43, 23, 21, 53, 64, 32, 40, 33, 25, 50]
sachin = [123, 32, 123, 33, 63, 44, 44, 32, 40, 45, 35, 39, 58]

# Setup
x = np.arange(len(years))  # positions on x-axis
width = 0.25               # width of each bar

# Plot
plt.figure(figsize=(14,6))
plt.bar(x - width, virat, width, label='Virat')
plt.bar(x, sehwag, width, label='Sehwag')
plt.bar(x + width, sachin, width, label='Sachin')

# Labels & Title
plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Players Performance Over the Years')
plt.xticks(x, years, rotation=45)  # set x-ticks as years
plt.legend()

# Show Plot
plt.show()
```
