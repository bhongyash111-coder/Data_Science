```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("iris.csv", sep="\t")  
df.head()
```

```python
df.info()
df.describe()
```

```python
sns.pairplot(df, hue="class", palette="Set2")
plt.show()
```

```python
sns.boxplot(x="class", y="sepallength", data=df, hue="class", palette="Set2", legend=False)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop(columns="class").corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation")
plt.show()
```
