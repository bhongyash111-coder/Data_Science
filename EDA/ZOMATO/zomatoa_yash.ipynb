```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

## observation

2. whenratinga re beteween 4.0 to 3.4 very good
3. when rating is between  3.5 to 3.9 avergage
4. when rating is between 3.0 to 3.4 average


```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','yellow','orange','green'])
```
